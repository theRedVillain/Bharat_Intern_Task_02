EDA of any genre :

```python
movie_df[movie_df['Action'] == 1]
```

```python
class EDA_Genre:
  final_lst = []
  def __init__(self, genre):
    self.genre = genre
    self.lst = {}
    self.votes_lst = {}


  def eda(self):
    temp_df = movie_df[movie_df[f'{self.genre}']==1]
    print(f"The Number of Directors in this genre {self.genre} are \n" , temp_df['Director'].nunique())
    print()
    print(f"The number of movies in this genre {self.genre} are \n" , temp_df.shape[0])
    print()
    print(f"Most Costliest Movies of this genre {self.genre} was \n" , temp_df.nlargest(1 , 'Revenue')['Title'] )
    print()
    print(f"Most Successful Movie of this genre {self.genre} is \n", temp_df.nlargest(1,'Rating')['Title'])
    print()

    print()
    print("The most sucessful directors in this genre are \n" , temp_df.groupby('Director',as_index = False).agg({
        'Votes':'mean'
    }).nlargest(10 , 'Votes'))
    print('------------/---------------------------/------------------------/---------')
    print()
    print('Year Wise Distribution on the popularity of the genre')
    self.YoY_Dist = temp_df.groupby('Year',as_index = False).agg({
        'Rank':'count',
        'Votes':'mean'
    })
    self.YoY_Dist.columns = ['Year' , 'Number of Movies Made','Average Votes']
    self.YoY_Dist.sort_values(by = ['Number of Movies Made','Average Votes'] , ascending= [False,False],inplace = True)
    print(self.YoY_Dist)
    self.YoY_Dist['Genre Type'] = self.genre
    print('----------/-----------------------/------------------------/-----------------')
    print()
    self.lst[f'{self.genre}'] = temp_df['Rating'].mean()
    self.votes_lst[f'{self.genre}'] = temp_df['Votes'].mean()
    print()
    EDA_Genre.final_lst.append(self.YoY_Dist[['Year','Number of Movies Made','Genre Type']])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
Y_new = le.fit_transform(Y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2 , random_state =11)
```

```python
print("The shape of X_Train is ",x_train.shape)
print("The shape of X_Test is ",x_test.shape)
print("The shape of Y_Train is ",y_train.shape)
print("The shape of Y_Test is ",y_test.shape)
```

```python
dt = DecisionTreeClassifier(criterion = 'gini' , min_samples_leaf = 8 , max_depth = 10 , random_state = 101)
```

```python
dt.fit(x_train,y_train)
```

```python
results = dt.predict(x_test)
```

```python
# Accuracy of the model

dt.score(x_test,y_test)
```

Confusion Matrix :

```python
print(confusion_matrix(y_test , results))
```

Classification Report :

```python
print(classification_report(y_test,results))
```
